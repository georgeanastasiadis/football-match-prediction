```python
df['Date'] = pd.to_datetime(df['Date'],infer_datetime_format=True, format='%d/%m/%Y')
```

```python
# Calculating the average of betting sites odd for each result
avgH = round((df['B365H'] + df['PSH'] + df['WHH'] + df['VCH'])/4,2)
avgD = round((df['B365D'] + df['PSD'] + df['WHD'] + df['VCD'])/4,2)
avgA = round((df['B365A'] + df['PSA'] + df['WHA'] + df['VCA'])/4,2)

df['AVGH'] = avgH
df['AVGD'] = avgD
df['AVGA'] = avgA

#drop the existing columns with betting odds
df = df.drop(['PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA', 'VCH', 
              'VCD', 'VCA', 'B365H', 'B365D', 'B365A'],axis=1)
```

```python
#transformation of result into numeric type data
def result(row):
    if row['FTR'] == 'H':
        val = 1
    elif row['FTR'] == 'D':
        val = 0
    else:
        val = 2
    return val

df['Result'] = df.apply(result,axis=1)
df = df.drop('FTR',axis=1)
```

```python
#Average of home and away goal depending stats of each team from fixture to fixture
df['HATT'] = df.groupby(['season','HomeTeam'])['FTHG'].apply(lambda x: x.shift().expanding().mean()).fillna(0)
df['HCONC'] = df.groupby(['season', 'HomeTeam'])['FTAG'].apply(lambda x: x.shift().expanding().mean()).fillna(0)
df['AATT'] = df.groupby(['season', 'AwayTeam'])['FTAG'].apply(lambda x: x.shift().expanding().mean()).fillna(0)
df['ACONC'] = df.groupby(['season', 'AwayTeam'])['FTHG'].apply(lambda x: x.shift().expanding().mean()).fillna(0)
```

```python
from copy import deepcopy

def match_result(home_team, Result):
    if Result == 1:
        if home_team:
            return 'W'
        else:
            return 'L'
    elif Result == 2:
        if home_team:
            return 'L'
        else:
            return'W'
    else:
        return 'D'

def last_5_games_form(df):
    teams = {key:['NG'] for key in df.HomeTeam.unique()} #NG since in the begining of the season there is no game history
  
    for i, row in df.iterrows():
        teams[row['HomeTeam']].append(match_result(home_team = True, Result = row['Result']))
        teams[row['AwayTeam']].append(match_result(home_team = False, Result = row['Result']))
    
  
    teams2 = deepcopy(teams)
    teams3 = deepcopy(teams)
    teams4 = deepcopy(teams)
    teams5 = deepcopy(teams)   
    
    for key in teams:
        teams2[key].insert(0,'NG')
    
        teams3[key].insert(0,'NG')
        teams3[key].insert(0,'NG')
    
        teams4[key].insert(0,'NG')
        teams4[key].insert(0,'NG')
        teams4[key].insert(0,'NG')
    
        teams5[key].insert(0,'NG')
        teams5[key].insert(0,'NG')
        teams5[key].insert(0,'NG')
        teams5[key].insert(0,'NG')

  # initialize last 5 games to 'M'
    for i in range(1, 6):
        df['HM' + str(i)] = 'NG'
        df['AM' + str(i)] = 'NG'
  
    for i, row in df.iterrows():
    
        df.at[i, 'HM5'] = teams5[row.HomeTeam].pop(0)
        df.at[i, 'HM4'] = teams4[row.HomeTeam].pop(0)
        df.at[i, 'HM3'] = teams3[row.HomeTeam].pop(0)
        df.at[i, 'HM2'] = teams2[row.HomeTeam].pop(0)
        df.at[i, 'HM1'] = teams[row.HomeTeam].pop(0)
    
        df.at[i, 'AM5'] = teams5[row.AwayTeam].pop(0)
        df.at[i, 'AM4'] = teams4[row.AwayTeam].pop(0)
        df.at[i, 'AM3'] = teams3[row.AwayTeam].pop(0)
        df.at[i, 'AM2'] = teams2[row.AwayTeam].pop(0)
        df.at[i, 'AM1'] = teams[row.AwayTeam].pop(0)
  
    return df
```

```python
s1213 = last_5_games_form(df.loc[df['season']=='2012/2013'])
s1314 = last_5_games_form(df.loc[df['season']=='2013/2014'])
s1415 = last_5_games_form(df.loc[df['season']=='2014/2015'])
s1516 = last_5_games_form(df.loc[df['season']=='2015/2016'])
s1617 = last_5_games_form(df.loc[df['season']=='2016/2017'])
s1718 = last_5_games_form(df.loc[df['season']=='2017/2018'])
s1819 = last_5_games_form(df.loc[df['season']=='2018/2019'])
s1920 = last_5_games_form(df.loc[df['season']=='2019/2020'])
s2021 = last_5_games_form(df.loc[df['season']=='2020/2021'])
s2122 = last_5_games_form(df.loc[df['season']=='2021/2022'])
```

```python
df = pd.concat([s1213 , s1314, s1415, s1516, s1617, s1718, s1819, s1920, s2021, s2122])
```

```python
df['HomeTeamForm'] = df['HM1'] + df['HM2'] + df['HM3'] + df['HM4'] + df['HM5']
df['AwayTeamForm'] = df['AM1'] + df['AM2'] + df['AM3'] + df['AM4'] + df['AM5']

df = df.drop(['HM1', 'HM2', 'HM3', 'HM4', 'HM5', 'AM1', 'AM2', 'AM3', 'AM4', 'AM5'], axis=1)
```
