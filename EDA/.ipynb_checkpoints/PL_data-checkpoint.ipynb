First we are going to plot some generic information of the Engalnd Premier League

```python
#figure resize
plt.figure(figsize=(10,6))

plt.plot(df['season'].unique(), df.groupby('season')['FTHG'].sum() + df.groupby('season')['FTAG'].sum(), marker = 'o')

plt.style.use('bmh')
plt.title('Total Goals per Season')
plt.xlabel('Seasons')
plt.ylabel('Total Goals')
```

```python
plt.figure(figsize=(12,8))

X_label = df['season'].unique()
X_axis = np.arange(len(df['season'].unique()))

plt.bar(X_axis - 0.2, df.groupby('season')['FTHG'].sum(),0.4, label = 'Home Goals')
plt.bar(X_axis + 0.2, df.groupby('season')['FTAG'].sum(),0.4, label = 'Away Goals')

plt.style.use('bmh')
plt.xticks(X_axis, X_label)
plt.title('Home vs Away Goals Per Season')
plt.xlabel('Seasons')
plt.ylabel('Goals')

plt.legend()
```

We can see that most of the time the sum of the number of goals for the home team is much higher than this of the away team. Only at season 2020-2021 the amount of goals for the home and away teams are almost the same, maybe that has to do with the absence of football fans due to covid measures. 

```python
plt.figure(figsize=(12,8))

X_label = df['season'].unique()
X_axis = np.arange(len(df['season'].unique()))

plt.bar(X_axis - 0.2, df.groupby('season')['Result'].apply(lambda x: (x == 1).sum()),0.4, label = 'Home Wins')
plt.bar(X_axis + 0.2, df.groupby('season')['Result'].apply(lambda x: (x == 2).sum()),0.4, label = 'Away Wins')

plt.style.use('bmh')
plt.xticks(X_axis, X_label)
plt.title('Home vs Away Wins Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Wins')

plt.legend()
```

During the season 2020/2021 where almost all matches had to be played behind closed doors, we observe
that the number of wins of away teams is higher than the number of wins of the home teams.

We want to check if the team with the best attack wins the EPL

```python
team_home_scored = df.groupby(['season','HomeTeam'])['FTHG'].sum()
team_home_scored = team_home_scored.rename_axis(['Season','Team Name'])

team_away_scored = df.groupby(['season','AwayTeam'])['FTAG'].sum()
team_away_scored = team_away_scored.rename_axis(['Season','Team Name'])


team_goals_scored = team_home_scored + team_away_scored

best_goalscoring_team = team_goals_scored.loc[team_goals_scored.groupby(level=0).idxmax()]

best_goalscoring_team
```

We can see that not always the team which scored the most goals wins the EPL.

```python
team_home_conceded = df.groupby(['season','HomeTeam'])['FTAG'].sum()
team_home_conceded = team_home_conceded.rename_axis(['Season','Team Name'])

team_away_conceded = df.groupby(['season','AwayTeam'])['FTHG'].sum()
team_away_conceded = team_away_conceded.rename_axis(['Season','Team Name'])

#Number of goals each team conceded per season
team_goal_conceded = team_home_conceded + team_away_conceded

best_team_at_defense = team_goal_conceded.loc[team_goal_conceded.groupby(level=0).idxmin()]
best_team_at_defense
```

```python
overall_stats = team_goals_scored - team_goal_conceded
best_overall_stats = overall_stats.loc[overall_stats.groupby(level=0).idxmax()]
best_overall_stats
```

6 out of 10 times the team with the best overall performance at wins the championship. 
So we need to take into consideration both the attacking and defensive performance when
we want to predict a match.

It is clearly that we have an unbalanced set of data. The number of home wins is much bigger than the number of the draws and the number of the away wins.

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='HST', y='FTHG', data=df, jitter = True, alpha=0.4)

plt.xticks(np.arange(0,22,2))
plt.yticks(np.arange(0,10,1))
plt.xlabel('Home Team Shots on Target')
plt.ylabel('Home Team Goals')
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='AST', y='FTAG', data=df, jitter = True, alpha=0.4)

plt.xticks(np.arange(0,22,2))
plt.yticks(np.arange(0,10,1))
plt.xlabel('Away Team Shots on Target')
plt.ylabel('Away Team Goals')
```

We can see that more shots on target leads to more goals, but that isn't always the case. There are a lot of times when the team with a high number of shots doesn't score a goal.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='AVGH', y='AVGA', data=df, alpha=0.2)

plt.xlabel('Avg Home Win Odds')
plt.ylabel('Avg Away Win Odds')
```

It is something that we have anticipated. The lower the home win odds, the bigger the away win odds. That has to do with the team dynamics.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='AVGA', y = 'AVGD', data = df ,alpha=0.2)

plt.xlabel('Away team odds')
plt.ylabel('Draw odds')
```

The average away team betting odd is higly correlated with the avarege draw odds. 

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Result', y='AVGH', data=df, jitter=True, size=3, alpha=0.3)

plt.title('Results based on home win odds')
plt.ylabel('Home win odds')
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Result', y='AVGA', data=df, jitter=True, size=3, alpha=0.3)

plt.title('Results based on away win odds')
plt.ylabel('Avg away win odds')
```

Take into consideration home win odd for the prediction of the result of a game seems to be a good practice. But because there is a possibility of having a draw I don't think it is a good practice.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='HATT', y='AVGA', data=df, alpha=0.3)

plt.xlabel('Average Home Attack')
plt.ylabel('Average Away Win Odds')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='AATT', y='AVGH', data=df, alpha=0.3)

plt.xlabel('Average Away Attack')
plt.ylabel('Average Home Win Odds')
```

We can see that the attacking power of the home and away team affects the price of the away and home win odds respectively. It is something that we have anticipated because it indicates the strength of the team.

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Result', y='AATT', data=df, jitter=True, size=3)

plt.ylabel('Avg Attacking power of away teams')
```

When the performance of the visiting team is weak it is highly possible that the home team is going to win. But as the attacking strength of the visitor is higher they will propably leave with a good result.
