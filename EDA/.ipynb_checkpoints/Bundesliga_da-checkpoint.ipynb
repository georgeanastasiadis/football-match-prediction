```python
s1213 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1213.csv')
s1314 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1314.csv')
s1415 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1415.csv')
s1516 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1516.csv')
s1617 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1617.csv')
s1718 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1718.csv')
s1819 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1819.csv')
s1920 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-1920.csv')
s2021 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-2021.csv')
s2122 = pd.read_csv(r'C:\Users\George\Desktop\Data Mining at Sports\Data\Fixtures\Bundesliga\D1-2122.csv')

s1213['season'] = '2012/2013'
s1314['season'] = '2013/2014'
s1415['season'] = '2014/2015'
s1516['season'] = '2015/2016'
s1617['season'] = '2016/2017'
s1718['season'] = '2017/2018'
s1819['season'] = '2018/2019'
s1920['season'] = '2019/2020'
s2021['season'] = '2020/2021'
s2122['season'] = '2021/2022'
```

```python
all_csv = pd.concat([s1213, s1314, s1415, s1516, s1617, s1718, s1819, s1920, s2021, s2122], ignore_index=True)
```

```python
df = all_csv[['season','Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTR',
              'HS', 'AS', 'HST','AST','HC','AC', 'HF', 'AF', 'B365H', 'B365A', 'B365D', 
              'BWH','BWD', 'BWA', 'IWH', 'IWD', 'IWA', 'PSH', 'PSD', 'PSA',
              'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA']]
```

```python
avgH = round((df['B365H'] + df['BWH'] + df['IWH'] + df['PSH'] + df['WHH'] + df['VCH'])/6,2)
avgD = round((df['B365D'] + df['BWD'] + df['IWD'] + df['PSD'] + df['WHD'] + df['VCD'])/6,2)
avgA = round((df['B365A'] + df['BWA'] + df['IWA'] + df['PSA'] + df['WHA'] + df['VCA'])/6,2)

df['AVGH'] = avgH
df['AVGD'] = avgD
df['AVGA'] = avgA

#drop the column with betting odds
df = df.drop(['BWH','BWD', 'BWA', 'IWH', 'IWD', 'IWA', 'PSH', 'PSD', 'PSA',
              'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA', 'B365H', 'B365D', 'B365A'],axis=1)
```

```python
plt.figure(figsize=(10,8))

plt.plot(df['season'].unique(), df.groupby('season')['FTHG'].sum()+df.groupby('season')['FTAG'].sum(), marker = 'o')

plt.style.use('bmh')
plt.title('Total Goals per Season')
plt.xlabel('Seasons')
plt.ylabel('Total Goals')
```

```python
plt.figure(figsize=(12,8))

X_label = df['season'].unique()
X_axis = np.arange(len(df['season'].unique()))

plt.bar(X_axis - 0.2, df.groupby('season')['FTHG'].sum(),0.4, label = 'Home Goals')
plt.bar(X_axis + 0.2, df.groupby('season')['FTAG'].sum(),0.4, label = 'Away Goals')

plt.style.use('bmh')
plt.xticks(X_axis, X_label)
plt.title('Home vs Away Goals Per Season')
plt.xlabel('Seasons')
plt.ylabel('Goals')

plt.legend()
```

We can see that the total number of goals that the teams achieve at home turf , is every season higher than the
total number of goals the teams achieve at away turf.

```python
plt.figure(figsize=(12,8))

X_label = df['season'].unique()
X_axis = np.arange(len(df['season'].unique()))

plt.bar(X_axis - 0.2, df.groupby('season')['FTR'].apply(lambda x: (x == 'H').sum()),0.4, label = 'Home Wins')
plt.bar(X_axis + 0.2, df.groupby('season')['FTR'].apply(lambda x: (x == 'A').sum()),0.4, label = 'Away Wins')

plt.style.use('bmh')
plt.xticks(X_axis, X_label)
plt.title('Home vs Away Wins Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Wins')

plt.legend()
```

```python
team_home_scored = df.groupby(['season','HomeTeam'])['FTHG'].sum()
team_home_scored = team_home_scored.rename_axis(['Season','Team Name'])

team_away_scored = df.groupby(['season','AwayTeam'])['FTAG'].sum()
team_away_scored = team_away_scored.rename_axis(['Season','Team Name'])


team_goals_scored = team_home_scored + team_away_scored

best_goalscoring_team = team_goals_scored.loc[team_goals_scored.groupby(level=0).idxmax()]

best_goalscoring_team
```

9 out of 10 time the team with the best attack won the championship.

```python
team_home_conceded = df.groupby(['season','HomeTeam'])['FTAG'].sum()
team_home_conceded = team_home_conceded.rename_axis(['Season','Team Name'])

team_away_conceded = df.groupby(['season','AwayTeam'])['FTHG'].sum()
team_away_conceded = team_away_conceded.rename_axis(['Season','Team Name'])

#Number of goals each team conceded per season
team_goal_conceded = team_home_conceded + team_away_conceded

best_team_at_defense = team_goal_conceded.loc[team_goal_conceded.groupby(level=0).idxmin()]
best_team_at_defense
```

8 out of 10 times the team with the best defence won the championship.

```python
overall_stats = team_goals_scored - team_goal_conceded
best_overall_stats = overall_stats.loc[overall_stats.groupby(level=0).idxmax()]
best_overall_stats
```

10 out of 10 times the team with the best overall stats won the championship. So we can imply that 
the the number of goal a team scores and concede play a significant role at the number of wins a team
has.

```python
sns.countplot(x='FTR', data=df)
```

We can see that our classification data set is imbalanced.

```python
#transformation of result into categorical type data
def result(row):
    if row['FTR'] == 'H':
        val = 1
    elif row['FTR'] == 'D':
        val = 0
    else:
        val = 2
    return val

df['Result'] = df.apply(result,axis=1)
df = df.drop('FTR',axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.drop(['Date', 'HomeTeam', 'AwayTeam','TGoal'],axis=1).corr(),annot=True, cmap='viridis')
```

```python
plt.figure(figsize=(10,6))
plt.style.use('bmh')
sns.stripplot(x='Result', y='FTAG', data=df, jitter=True, size=3)
```

We can see that when the away team scores 2 or more goals it is highly possible that 
it is going to win or at least achieve a draw.

```python
plt.figure(figsize=(10,6))
plt.style.use('bmh')
sns.scatterplot(x='AVGA', y='AVGD', data=df)
```

We can see that the the average draw odds(AVGD) and the average away team win odds (AVGA)
have a positive association because as AVGA increases so does AVGD. 


```python
plt.figure(figsize=(10,6))
plt.style.use('bmh')
sns.scatterplot(x='HS', y='HST', data=df)
```

It is something that we have anticipated The higher the number of shots,
the higher the amount of shots that find the target. The same thing stands
for the away team.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='AVGH', y='AVGA', data=df)

plt.xticks(np.arange(0,26,2))
plt.yticks(np.arange(0,40,2))

plt.xlabel('Average Home Win Odds')
plt.ylabel('Average Away Win Odds')
plt.show()
```

It is something that we have anticipated. The lower the average home win odd is the higher the average away win odd.

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='FTHG', y='HST', data=df, jitter=True, size=3)
```

We can see that when the home team does a lot shots that find the target it is possible that they will score. But that doesn't mean that the number of shots and goals are going to be proportional.
