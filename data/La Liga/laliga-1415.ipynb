```python
import numpy as np
import pandas as pd
```

```python
s1415 = pd.read_csv(r'C:\Users\George\Desktop\football-match-prediction-using-ml-techniques\data\La Liga\SP1-1415.csv')

understats = pd.read_csv(r'C:\Users\George\Desktop\football-match-prediction-using-ml-techniques\data\understat_per_game.csv')
```

```python
s1415 = s1415[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG','HTHG','HTAG', 'FTR',
              'HS', 'AS', 'HST','AST','HC','AC', 'HF', 'AF', 'B365H', 'B365A', 'B365D',
               'PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA']].reset_index(drop=True)
```

```python
s1415.head()
```

```python
s1415['Date'] = pd.to_datetime(s1415['Date'],format="%d/%m/%y")
```

```python
understats['Date'] = pd.to_datetime(understats['date']).dt.strftime("%Y-%m-%d")
understats['Date'] = pd.to_datetime(understats['Date'])
```

```python
understats =understats.replace({'team':{'Malaga':'Malaga', 'Sevilla':'Sevilla',
       'Deportivo La Coruna':'La Coruna', 'Real Sociedad':'Sociedad', 'Espanyol':'Espanol', 'Getafe':'Getafe',
       'Atletico Madrid':'Ath Madrid', 'Rayo Vallecano':'Vallecano', 'Valencia':'Valencia', 'Athletic Club':'Ath Bilbao',
       'Barcelona':'Barcelona', 'Real Madrid':'Real Madrid', 'Levante':'Levante', 'Celta Vigo':'Celta', 'Villarreal':'Villarreal',
       'Granada':'Granada', 'Eibar':'Eibar', 'Cordoba':'Cordoba', 'Elche':'Elche', 'Almeria':'Almeria', 'Las Palmas':'Las Palmas',
       'Sporting Gijon':'Gijon', 'Real Betis':'Betis', 'Osasuna':'Osasuna', 'Alaves':'Alaves', 'Leganes':'Leganes',
       'Girona':'Girona', 'Real Valladolid':'Valladolid', 'SD Huesca':'Huesca', 'Mallorca':'Mallorca'}})
```

```python
df = pd.merge(s1415,understats, how='left',left_on=['Date','HomeTeam'],right_on=['Date','team'])
```

```python
df.isnull().sum()
```

```python
#We know that most of the times the books odds are almsot the same. So we are going to replace the null values with the
#average of the 3 other sites odds
index = df.loc[df['PSH'].isnull()].index
for row in index:
    df.at[row,'PSH'] = (df.loc[row,'B365H'] + df.loc[row,'WHH'] + df.loc[row,'VCH'])/3
    df.at[row,'PSD'] = (df.loc[row,'B365D'] + df.loc[row,'WHD'] + df.loc[row,'VCD'])/3
    df.at[row,'PSA'] = (df.loc[row,'B365A'] + df.loc[row,'WHA'] + df.loc[row,'VCA'])/3
    
df['PSH'] = df['PSH'].round(2)
df['PSA'] = df['PSA'].round(2)
df['PSD'] = df['PSD'].round(2)
```

```python
# Calculating the average of betting sites odd for each result
avgH = round((df['B365H'] + df['PSH'] + df['WHH'] + df['VCH'])/4,2)
avgD = round((df['B365D'] + df['PSD'] + df['WHD'] + df['VCD'])/4,2)
avgA = round((df['B365A'] + df['PSA'] + df['WHA'] + df['VCA'])/4,2)

df['AVGH'] = avgH
df['AVGD'] = avgD
df['AVGA'] = avgA

#drop the existing columns with betting odds
df = df.drop(['PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA', 'VCH', 
              'VCD', 'VCA', 'B365H', 'B365D', 'B365A'],axis=1)
```

```python
df['season'] = '2014/2015'
```

```python
df.rename(columns = {'xG':"HxG", 'xGA':'AxG','deep':'Hdeep','deep_allowed':'Adeep',
                     'ppda_coef':'Hppda_coef','oppda_coef':'Appda_coef'},inplace=True)

df = df.drop(['npxG','npxGA','scored', 'missed', 'xpts', 'result', 'wins',
              'draws', 'loses', 'pts', 'npxGD', 'ppda_att', 'ppda_def',
              'oppda_att', 'oppda_def', 'team', 'xG_diff', 'xGA_diff',
              'xpts_diff', 'date', 'year','h_a'],axis=1)
```
